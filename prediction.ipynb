```python
import numpy as np
import pandas as pd
```

```python
Team1_bat = pd.read_csv("./data/Pak-batting.csv")
Team2_bat = pd.read_csv("./data/Afg-batting.csv")
```

```python
Team1_bowl = pd.read_csv("./data/Pak-bowling.csv")
Team2_bowl = pd.read_csv("./data/Afg-bowling.csv")
```

```python
Team1_bat
```

```python
batting_scores = []
for i in range(Team1_bat.shape[0]):
    u = np.sqrt(Team1_bat.iloc[i]['Matches']/Team1_bat.iloc[i]["Innings"])
    v = 20*Team1_bat.iloc[i]["200"]+10*Team1_bat.iloc[i]["100"]+5*Team1_bat.iloc[i]["50"]
    w = 0.3 * v + 0.7 * Team1_bat.iloc[i]["Avg"]
    y = Team1_bat.iloc[i]["SR"]
#     x = 0.2 * (Team1_bat.iloc[i]["Innings"]/Team1_bat.iloc[i]["4s"] + Team1_bat.iloc[i]["Innings"]/Team1_bat.iloc[i]["6s"])
#     if np.isinf(x):
#         career_score = u*w*y
#     else:
#         career_score = u*w*y*x
    career_score = u*w*y
    batting_scores.append(career_score)
    
for i in range(Team2_bat.shape[0]):
    u = np.sqrt(Team2_bat.iloc[i]['Matches']/Team2_bat.iloc[i]["Innings"])
    v = 20*Team2_bat.iloc[i]["200"]+10*Team2_bat.iloc[i]["100"]+5*Team2_bat.iloc[i]["50"]
    w = 0.3 * v + 0.7 * Team2_bat.iloc[i]["Avg"]
    y = Team2_bat.iloc[i]["SR"]
#     x = 0.2*(Team2_bat.iloc[i]["Innings"]/Team2_bat.iloc[i]["4s"])  + 0.7*(Team2_bat.iloc[i]["Innings"]/Team2_bat.iloc[i]["6s"])
#     if np.isinf(x):
#         career_score = u*w*y
#     else:
#         career_score = u*w*y*x
    career_score = u*w*y
    batting_scores.append(career_score)
```

```python
m = max(batting_scores)
for i in range(len(batting_scores)):
    batting_scores[i] = batting_scores[i]/m
```

```python
bowling_scores = []
for i in range(Team1_bowl.shape[0]):
    u = np.sqrt(Team1_bowl.iloc[i]["Innings"]/Team1_bowl.iloc[i]["Matches"])
    v = 10*Team1_bowl.iloc[i]["5W"]+Team1_bowl.iloc[i]["Wickets"]
    w = Team1_bowl.iloc[i]["Avg"] * Team1_bowl.iloc[i]["Econ"] * Team1_bowl.iloc[i]["SR"]
    y = Team1_bowl.iloc[i]["Bowls"]/Team1_bowl.iloc[i]["Runs"]
#     z = Team1_bowl.iloc[i]["SR"]
#     if z>0:
#         bowler_score = u*(v/w)*(v/z)+y
#     else:
#         bowler_score = u*v/w+y
    bowler_score = u*v/w+y
    bowling_scores.append(bowler_score)
    
for i in range(Team2_bowl.shape[0]):
    u = np.sqrt(int(Team2_bowl.iloc[i]["Innings"])/Team2_bowl.iloc[i]["Matches"])
    v = 10*Team2_bowl.iloc[i]["5W"]+Team2_bowl.iloc[i]["Wickets"]
    w = Team2_bowl.iloc[i]["Avg"] * Team2_bowl.iloc[i]["Econ"] * Team2_bowl.iloc[i]["SR"]
    y = Team2_bowl.iloc[i]["Bowls"]/Team2_bowl.iloc[i]["Runs"]
#     z = Team2_bowl.iloc[i]["SR"]
#     if z>0:
#         bowler_score = u*(v/w)*(v/z)+y
#     else:
#         bowler_score = u*v/w+y
    bowler_score = u*v/w+y
    bowling_scores.append(bowler_score)
```

```python
m = np.nanmax(bowling_scores)
for i in range(len(bowling_scores)):
    bowling_scores[i] = bowling_scores[i]/m
```

```python
batting_scores = np.array(batting_scores)
bowling_scores = np.array(bowling_scores)
```

```python
bowling_scores[np.isnan(bowling_scores)] = 0
batting_scores[np.isnan(batting_scores)] = 0
```

```python
Team1_batting = np.sum(batting_scores[:15])
Team1_bowl = np.sum(bowling_scores[:15])
Team2_batting = np.sum(batting_scores[15:])
Team2_bowl = np.sum(bowling_scores[15:])
```
